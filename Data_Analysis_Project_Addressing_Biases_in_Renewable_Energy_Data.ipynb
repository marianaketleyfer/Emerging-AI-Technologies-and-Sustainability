```python
def print_section_title(title):
    print(colored(title, 'red', attrs=['bold', 'underline']))

def display_head_and_tail(dataframe, head=10):
    display(dataframe.head(head).style.set_caption("Head"))
    display(dataframe.tail(head).style.set_caption("Tail"))

def display_na(dataframe):
    na_df = dataframe.isnull().sum().reset_index()
    na_df.columns = ['Column', 'Number of NA']
    display(na_df.style.set_caption("Number of NA Values"))

def display_quantiles(dataframe):
    quantiles_df = dataframe.describe([0, 0.05, 0.50, 0.95, 0.99, 1]).T
    display(quantiles_df.style.format("{:.2f}").set_caption("Quantiles"))

def check_df(dataframe, head=5):
    print_section_title('Shape')
    print(dataframe.shape)
    print_section_title('Types')
    print(dataframe.dtypes.to_frame('Data Type').style.set_caption("Data Types"))
    print_section_title('Head & Tail')
    display_head_and_tail(dataframe, head)
    print_section_title('NA Values')
    display_na(dataframe)
    print_section_title('Quantiles')
    display_quantiles(dataframe)

check_df(df_1)
```

```python
#30 Countries/Regions for Renewable Energy Share in the Last 10 Years
latest_10_years = df_1['Year'].unique()[-10:]

filtered_data = df_1[df_1['Year'].isin(latest_10_years)]

top_30_per_year = filtered_data.groupby('Year').apply(lambda x: x.nlargest(30, 'Renewables (% equivalent primary energy)'))

pivot_data = top_30_per_year.pivot(index='Entity', columns='Year', values='Renewables (% equivalent primary energy)').fillna(0)

sorted_pivot_data = pivot_data.sum(axis=1).sort_values(ascending=False)

sorted_pivot_data = pivot_data.loc[sorted_pivot_data.index]

fig, ax = plt.subplots(figsize=(30, 10))
sorted_pivot_data.plot(kind='bar', stacked=True, ax=ax, colormap='turbo', edgecolor='white', width=0.6)
ax.set_title('30 Countries/Regions for Renewable Energy Share in the Last 10 Years', fontsize=18, fontweight='bold')
ax.set_xlabel('Country/Region', fontsize=14, fontweight='bold')
ax.set_ylabel('Renewable Energy Share (%)', fontsize=14, fontweight='bold')
ax.legend(title='Year', fontsize=10)
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
#20 Countries/Regions with the Lowest Renewable Energy Share in the Last 6 Years
latest_10_years = df_1['Year'].unique()[-10:]

heatmap_data = pd.DataFrame()

for year in latest_10_years:
    specific_year_data = df_1[df_1['Year'] == year]
    bottom_countries_year = specific_year_data.nsmallest(20, 'Renewables (% equivalent primary energy)')
    heatmap_data[year] = bottom_countries_year.set_index('Entity')['Renewables (% equivalent primary energy)']

title_font_size = 16
label_font_size = 12

plt.figure(figsize=(14, 10))
sns.heatmap(heatmap_data, cmap='Spectral', annot=True, linewidths=0.5, linecolor='gray', cbar_kws={'label': 'Renewable Energy Share (%)'})
plt.title('Renewable Energy Share in Bottom 20 Countries/Regions', fontsize=title_font_size, fontweight='bold')
plt.xlabel('Year', fontsize=label_font_size, fontweight='bold')
plt.ylabel('Country/Region', fontsize=label_font_size, fontweight='bold')
plt.show()
```

```python
#Re-sampling: Adjust the dataset to balance representation

from sklearn.utils import resample

entity_counts = df_1['Entity'].value_counts()
most_frequent_region = entity_counts.index[0]
least_frequent_region = entity_counts.index[-1]

df_majority = df_1[df_1['Entity'] == most_frequent_region]
df_minority = df_1[df_1['Entity'] == least_frequent_region]

df_minority_upsampled = resample(df_minority,
                                 replace=True,
                                 n_samples=len(df_majority),
                                 random_state=123)

df_upsampled = pd.concat([df_majority, df_minority_upsampled])

print(df_upsampled['Entity'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df_1 = pd.read_csv('/content/01 renewable-share-energy.csv')

df_upsampled_counts = df_upsampled

fig, ax = plt.subplots(1, 2, figsize=(16, 8))

sns.histplot(df_1['Renewables (% equivalent primary energy)'], kde=True, ax=ax[0], color='turquoise')
ax[0].set_title('Before Bias Mitigation', fontsize=16)
ax[0].set_xlabel('Renewable Energy Consumption', fontsize=14)
ax[0].set_ylabel('Frequency', fontsize=14)
ax[0].tick_params(axis='both', which='major', labelsize=12)
ax[0].grid(True, linestyle='--', alpha=0.7)

sns.histplot(df_upsampled_counts, kde=True, ax=ax[1], color='pink')
ax[1].set_title('After Bias Mitigation', fontsize=16)
ax[1].set_xlabel('Number of Entities', fontsize=14)
ax[1].set_ylabel('Frequency', fontsize=14)
ax[1].tick_params(axis='both', which='major', labelsize=12)
ax[1].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

encoder = OneHotEncoder(handle_unknown='ignore')
encoded_entities = encoder.fit_transform(df_1[['Entity']])

encoded_df = pd.DataFrame.sparse.from_spmatrix(encoded_entities, columns=encoder.get_feature_names_out(['Entity']))

X = pd.concat([encoded_df, df_1[['Renewables (% equivalent primary energy)']]], axis=1)
y = df_1['Renewables (% equivalent primary energy)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model_before = LinearRegression()
model_before.fit(X_train, y_train)
predictions_before = model_before.predict(X_test)
mse_before = mean_squared_error(y_test, predictions_before)

encoded_upsampled_entities = encoder.transform(df_upsampled[['Entity']])
encoded_upsampled_df = pd.DataFrame.sparse.from_spmatrix(encoded_upsampled_entities, columns=encoder.get_feature_names_out(['Entity']))

encoded_upsampled_df = encoded_upsampled_df.reset_index(drop=True)
df_upsampled_numeric = df_upsampled[['Renewables (% equivalent primary energy)']].reset_index(drop=True)
X_upsampled = pd.concat([encoded_upsampled_df, df_upsampled_numeric], axis=1)
y_upsampled = df_upsampled['Renewables (% equivalent primary energy)']  # Assuming this is your target variable

X_train_up, X_test_up, y_train_up, y_test_up = train_test_split(X_upsampled, y_upsampled, test_size=0.3, random_state=42)

model_after = LinearRegression()
model_after.fit(X_train_up, y_train_up)
predictions_after = model_after.predict(X_test_up)
mse_after = mean_squared_error(y_test_up, predictions_after)

print(f'MSE before bias mitigation: {mse_before}')
print(f'MSE after bias mitigation: {mse_after}')
```
